```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)

plt.plot(x,x,label='linear')
plt.plot(x,np.sin(x),label='linear')
plt.plot(x,x**3,label='linear')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()
plt.show()
```
